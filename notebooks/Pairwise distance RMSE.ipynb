```python
import dendropy
```

```python
t = "[&R] ((0:0.0360971597765934,1:3.339535381892265):0.0360971597765934,(2:0.0360971597765934,3:3.339535381892265):0.0360971597765934);"
true_tree = dendropy.Tree.get(data=t, schema="newick")
```

```python
import os, glob

pair = [0, 1]
tree_dict = dict()
pdm_dict = dict()
for k in (20, 30, 40, 50, 100, 200, 300, 400, 500, 1000, 5000):
    outdir = "/Users/gillianchu/raphael/repos/problin/results_estbl/m10_k{0}_estbl".format(k)
    for filename in glob.iglob(outdir+"/*"):
        # print(filename)
        est_tree = dendropy.Tree.get(path=filename, schema="newick")
        pdm = est_tree.phylogenetic_distance_matrix()
        if k not in tree_dict.keys():
            tree_dict[k] = []
            pdm_dict[k] = []
        tree_dict[k].append(est_tree)
        pdm_dict[k].append(pdm)
        
    
```

```python
import numpy as np

x = []
y = dict()

for k in (20, 30, 40, 50, 100, 200, 300, 400, 500, 1000, 5000):
    pair_dict = dict()
    for i in range(len(tree_dict[k])):
        t = tree_dict[k][i]
        pdm = pdm_dict[k][i]
        # print(t)
        for idx1, taxon1 in enumerate(t.taxon_namespace):
            for taxon2 in t.taxon_namespace:
                weighted_patristic_distance = pdm.patristic_distance(taxon1, taxon2)
                pair = (taxon1.label, taxon2.label)
                if pair not in pair_dict.keys():
                    pair_dict[pair] = []
                pair_dict[pair].append(weighted_patristic_distance)

    true_pdm = true_tree.phylogenetic_distance_matrix()
    for idx1, taxon1 in enumerate(true_tree.taxon_namespace):
            for taxon2 in true_tree.taxon_namespace:
                true_dist = true_pdm.patristic_distance(taxon1, taxon2)
                pair = (taxon1.label, taxon2.label)

                # compute the error between the true_dist and the new
                rmse = np.sqrt(((np.array([true_dist]*len(pair_dict[pair])) - np.array(pair_dict[pair]))**2).mean())
                # print(pair, rmse)
                if pair not in y:
                    y[pair] = []
                y[pair].append(rmse)
    x.append(k)
```

```python
import matplotlib.pyplot as plt

for pair in pair_dict.keys():
    plt.title("Length of Sequence (k) vs {0} Distance RMSE".format(pair))
    plt.plot(x, y[pair])
    ax = plt.gca()
    ax.set_ylim([0,12])
    plt.xlabel("k")
    plt.ylabel("Pairwise Distance RMSE")
    plt.savefig('/Users/gillianchu/raphael/repos/problin/results_estbl/plots/pair_{0}_{1}.jpg'.format(pair[0], pair[1]))
    plt.show()
```
